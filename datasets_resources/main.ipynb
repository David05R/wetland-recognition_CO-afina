### Adding more info.
From colombian_population_density.ipynb notebook in the folder, we extracted information to add in a new column. Starting with population density...


```python
# Read the population density file
df_density = pd.read_csv('col_population_density.csv')

# Read the Ramsar sites file
colWetlands = pd.read_csv('colWetlands.csv')

# Create a dictionary of population density by department
density_dict = dict(zip(df_density['Department'], df_density['Population Density (people/km2)']))

# Function to get population density based on the department
def get_density(row):
    # Extract the department from the 'large administrative region' column
    department = row['large administrative region'].split(',')[0].strip()
    
    # Handle special cases
    if department == 'Bogotá Distrito Capital':
        department = 'Bogotá'
    elif department == 'Municipios de Puerto Nariño y Leticia':
        department = 'Amazonas'
    
    # Look up the density in the dictionary
    return density_dict.get(department, 'Not available')

# Apply the function to create the new column and update the same document
colWetlands['Population Density (people/km2)'] = colWetlands.apply(get_density, axis=1)

# Save the result to a new CSV file
# colWetlands.to_csv('colWetlands_v2.csv', index=False)


```

### Another column.
This one will filter colombian Ramsar's sites following its Zone Type based in their conditions.

```python
def zone_characterization(row):
    # List of protected area conditions
    protected_area_conditions = [
        'National conservation designation',
        'Global international legal designations',
        'Regional international legal designations'
    ]
    
    # List of urban area keywords
    urban_area_keywords = [
        'Human settlements', 'Housing', 'Tourism', 'Commercial and industrial areas',
        'Household sewage', 'Housing and urban areas', 'Human settlements (non agricultural)',
        'Recreational and tourism activities', 'Roads and railroads', 'Tourism and recreation areas',
        'Transportation and service corridors', 'Unspecified development', 'urban waste water'
    ]
    
    # Check for protected area conditions
    if any(pd.notna(row[cond]) for cond in protected_area_conditions):
        return 'Protected Area'
    
    # Check for urban area conditions
    if any(keyword in str(row['Threats']) for keyword in urban_area_keywords):
        return 'Urban Area'
    
    # Default classification
    return 'Other'
```

```python
colWetlands['Zone type'] = colWetlands.apply(zone_characterization, axis=1)
colWetlands
```

#### Finally, we added a new last column.
We made a last column also related to the characterization of the area, this time, focused on the population density already calculated, defining whether it is low, medium or high density.

```python
# Function to classify zones by population density
def zone_characterization_by_population_density(row):
    density = row['Population Density (people/km2)']
    
    # Check if the density is a number
    if pd.isna(density) or density == 'Not available':
        return 'No data'
    
    density = float(density)
    
    if density < 40:
        return 'Low Density'
    elif 40 <= density <= 58.04:
        return 'Medium Density'
    else:  # density > 58.04
        return 'High Density'

# Apply the function to create the new classification column
colWetlands['Population Density Classification'] = colWetlands.apply(zone_characterization_by_population_density, axis=1)

# Save the updated DataFrame to a new CSV file
# colWetlands.to_csv('colWetlandsPro.csv', index=False)
```
