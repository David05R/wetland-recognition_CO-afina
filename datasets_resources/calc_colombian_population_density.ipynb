# About this code...
All this information was extracted from the web pages of the DANE and the Colombian Departments, respectively, population and surface area. The purpose of this code is to calculate the population density of the country as shown below.


```python
import pandas as pd

# Surface area data (km2)
surface_area_data = {
    'Department': ['Amazonas', 'Vichada', 'Caquetá', 'Meta', 'Guainía', 'Antioquía', 'Vaupés', 'Guaviare', 'Chocó', 'Casanare', 
                   'Nariño', 'Santander', 'Cauca', 'Bolívar', 'Córdoba', 'Putumayo', 'Arauca', 'Tolima', 'Boyacá', 'Magdalena', 
                   'Cesar', 'Cundinamarca', 'Valle del Cauca', 'Norte de Santander', 'La Guajira', 'Huila', 'Sucre', 'Caldas', 
                   'Risaralda', 'Atlántico', 'Quindió', 'San Andrés y Providencia', 'Bogotá'],
    'Surface_km2': [109665, 100242, 88965, 85635, 72238, 63612, 54135, 53460, 46530, 44640, 
                    33268, 30537, 29308, 25978, 25020, 24885, 23818, 23562, 23189, 23188, 
                    22905, 22633, 22140, 21658, 20848, 19890, 10917, 7888, 4140, 3388, 
                    1845, 52, 1587]
}

# Population data
population_data = {
    'Department': ['Bogotá', 'Antioquia', 'Valle del Cauca', 'Cundinamarca', 'Atlántico', 'Santander', 'Bolívar', 'Córdoba', 'Nariño', 'Norte de Santander', 
                   'Cauca', 'Magdalena', 'Tolima', 'Cesar', 'Boyacá', 'Huila', 'Meta', 'Caldas', 'La Guajira', 'Sucre', 
                   'Risaralda', 'Chocó', 'Quindío', 'Casanare', 'Caquetá', 'Putumayo', 'Arauca', 'Vichada', 'Guaviare', 'Amazonas', 
                   'San Andrés y Providencia', 'Guainía', 'Vaupés'],
    'Population': [7968792, 7088795, 4638029, 3445327, 2803565, 2357127, 2247283, 1898911, 1699570, 1696740, 
                   1558045, 1496163, 1374384, 1373581, 1298800, 1178453, 1130085, 1040284, 1038397, 994060, 
                   972304, 595138, 563076, 467775, 425053, 383042, 313097, 123304, 97616, 85056, 
                   62269, 56551, 46777]
}

# Create DataFrames
df_surface = pd.DataFrame(surface_area_data)
df_population = pd.DataFrame(population_data)

# Merge DataFrames by Department
df = pd.merge(df_surface, df_population, on='Department')

# Calculate population density
df['Population Density (people/km2)'] = df['Population'] / df['Surface_km2']

# Display result
print(df[['Department', 'Surface_km2', 'Population', 'Population Density (people/km2)']])


#df.to_csv("col_population_density.csv", index=False) #This line will create a file with the results (it is already in the folder ;))

```
